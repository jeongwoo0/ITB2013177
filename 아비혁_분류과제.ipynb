# 의사결정나무 모델

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
dt_clf = DecisionTreeClassifier()
X = health_sample_drop_row.loc[:, ['식전혈당(공복혈당)', '감마지티피']]
y = health_sample_drop_row.loc[:, ['rating']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.24, random_state=47)

tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train.values.ravel())

print("학습용 데이터 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("시험용 데이터 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
import graphviz

export_graphviz(tree, out_file="tree.dot", class_names=[ "50kg 미만", "70kg 미만", "90kg 미만", "110kg 미만", "130kg 이하"],
                feature_names=X.columns, impurity=False, filled=True)

with open("tree.dot") as f:
  dot_graph = f.read()

display(graphviz.Source(dot_graph))
```

# 앙상블 모델

## 랜덤 포레스트

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=131, random_state=0)
model = forest.fit(X_train, y_train.values.ravel())

print("학습용 데이터 세트 정확도: {:.3f}".format(forest.score(X_train, y_train)))
print("시험용 데이터 세트 정확도: {:.3f}".format(forest.score(X_test, y_test)))
```

## 그래디언트 부스팅

```python
from sklearn.ensemble import GradientBoostingClassifier

gbrt = GradientBoostingClassifier(random_state=0, max_depth=10, learning_rate=0.05)                           
gbrt.fit(X_train, y_train.values.ravel())

print("학습용 데이터 세트 정확도: {:.3f}".format(gbrt.score(X_train, y_train)))
print("시험용 데이터 세트 정확도: {:.3f}".format(gbrt.score(X_test, y_test)))
```

# 분류 모델 평가

## 교차 검증

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris


dt_scores = cross_val_score(tree, X_train, y_train.values.ravel(), cv=10, scoring='accuracy')  
rf_scores = cross_val_score(forest, X_train, y_train.values.ravel(), cv=10, scoring='accuracy')
gbrt_scores = cross_val_score(gbrt, X_train, y_train.values.ravel(), cv=10, scoring='accuracy')

print("Accuracy")
print("Decision tree: ", dt_scores) 
print("Random forest: ", rf_scores) 
print("Gradient boosting: ", gbrt_scores) 

print("Accuracy mean")
print("Decision tree :{:.3f}".format(dt_scores.mean())) 
print("Random forest :{:.3f}".format(rf_scores.mean())) 
print("Gradient boosting :{:.3f}".format(gbrt_scores.mean())) 
```

```python
from sklearn.model_selection import StratifiedKFold
kf = StratifiedKFold(n_splits=10)

print(kf)

dt_scores = cross_val_score(tree, X_train, y_train.values.ravel(), cv=kf, scoring='accuracy')  
rf_scores = cross_val_score(forest, X_train, y_train.values.ravel(), cv=kf, scoring='accuracy')
gbrt_scores = cross_val_score(gbrt, X_train, y_train.values.ravel(), cv=kf, scoring='accuracy')

print("Accuracy mean -- StratifiedKFold 이용")
print("Decision tree :{:.3f}".format(dt_scores.mean())) 
print("Random forest :{:.3f}".format(rf_scores.mean())) 
print("Gradient boosting :{:.3f}".format(gbrt_scores.mean())) 
```
